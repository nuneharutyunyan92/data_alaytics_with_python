 Part 1. 
   #  1.1 revenues including tax

```python
import pandas as pd
revenues =  [120, 250, 80, 310, 95]
    
df= pd.DataFrame({"revenue": revenues})
t = 0.18
df['revenue_including_tax'] = df["revenue"] * (1+t)
print(df)

```

# 1.2 final revenues after tax and discount


```python
import pandas as pd
revenues =  [120, 250, 80, 310, 95]
        
df= pd.DataFrame({"r": revenues})
df['final_revenue'] = df["r"] * (1+0.18) * (1-0.10)
print(df)
```

# 1.3. Boolean list indicating whether revenue > 100.

```python
df["high_revenue"] = df["final_revenue"] > 100
print(df)
```

```python
df["final_revenue"] > 100
```

# 1.4. Boolean list indicating whether revenue is between 100 and 300.

```python
df["mid_revenue"] = (df["final_revenue"] >= 100) & (df["final_revenue"] <= 300)
print(df)
```
